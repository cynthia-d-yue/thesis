```python
import pandas as pd
```

```python
df_festival = pd.read_csv('./data/festivals.csv', keep_default_na=False)
df_festival.head()
```

```python
df_film = pd.read_csv('./data/films.csv', keep_default_na=False)
df_film.head()
```

```python
df_film_festival = df_festival.merge(df_film, on=['id'], how='inner')
df_film_festival.head()
```

```python
df_film_festival.to_csv('./data/film_festivals.csv', index = False)
```
