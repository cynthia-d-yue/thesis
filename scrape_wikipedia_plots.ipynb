```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
df_film_festival = pd.read_csv('./data/film_festivals.csv', keep_default_na=False)
df_film_festival.head()
```

```python
plot_wikipedia = []

i = 0 # debugging
for link in df_film_festival['link_film']:
    print(i) # debugging
    print(link) # debugging
    
    if link == 'N/A':
        plot_wikipedia.append('N/A')
        print(link) # debugging
        print('\n') # debugging
        i += 1
        continue
    
    url = link
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'lxml')

    plot = ''
    paragraph = 1
    if soup.find('span', {'id': 'Plot'}):
        next_node = soup.find('span', {'id': 'Plot'}).parent.find_next_sibling('p')
        if next_node == None:
            plot = 'N/A'
            paragraph = 0
        print('found plot') # debugging
        print(next_node) # debugging
    else:
        plot = 'N/A'
        paragraph = 0

    while paragraph == 1:
        plot += next_node.get_text() + ' '
        next_node = next_node.next_sibling
        if next_node.name != 'p':
            paragraph = 0
    
    plot_wikipedia.append(plot)
    print('\n') # debugging
    i += 1
```

```python
df_film_festival['Plot_Wikipedia'] = plot_wikipedia
df_film_festival_with_plots = df_film_festival[~((df_film_festival['Plot'].isin(['', 'N/A'])) & (df_film_festival['Plot_Wikipedia'] == 'N/A'))].reset_index(drop=True)
```

```python
def select_plot(plot_api, plot_wikipedia):
    if len(plot_wikipedia) > len(plot_api):
        return plot_wikipedia
    else:
        return plot_api
    
def combine_plot(plot_api, plot_wikipedia):
    if plot_api in ['', 'N/A']:
        return plot_wikipedia
    elif plot_wikipedia == 'N/A':
        return plot_api
    else:
        return plot_wikipedia + plot_api
```

```python
df_film_festival_with_plots['Plot_Selected'] = df_film_festival_with_plots.apply(lambda row: select_plot(row['Plot'], row['Plot_Wikipedia']), axis=1) # selects the longer plot between API and Wikipedia
df_film_festival_with_plots['Plot_Combined'] = df_film_festival_with_plots.apply(lambda row: combine_plot(row['Plot'], row['Plot_Wikipedia']), axis=1) # combines plot from API and Wikipedia
```
